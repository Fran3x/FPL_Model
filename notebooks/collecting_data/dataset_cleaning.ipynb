# Expected points

```python
# FWD
df.loc[df["FPL_pos"] == "FWD", "xP"] = (df["Min"] > 0) + (df["Min"] >= 60) + df["xG"] * 4 + df["xA"] * 3 + df['Bonus']
```

```python
df.loc[df["FPL_pos"] == "MID", "xP"] = (df["Min"] > 0) + (df["Min"] >= 60) + df["xG"] * 5 + df["xA"] * 3 + df["Player_CS"] + df['Bonus']
```

```python
df.loc[df["FPL_pos"] == "DEF", "xP"] = (df["Min"] > 0) + (df["Min"] >= 60) + df["xG"] * 6 + df["xA"] * 3 + df["Player_CS"] * 4 + df['Bonus']
```

```python
df.loc[df["FPL_pos"] == "GK", "xP"] = df["FPL_points"]
```

# Summary & save

 <strong>0. Date </strong>- YYYY-MM-DD format 
 <strong>1. Day  </strong>- day of the week 
 <strong>2. Venue </strong> - home/away 
 <strong>3. Team  </strong> - team name
 <strong>4. Opponent   </strong> - opposite team name 
 <strong>5. Name       </strong> - player name
 <strong>6. Start         </strong> - Y:starting eleven, N:came on from bench, Y*:starting eleven - team captain
 <strong>7. Pos           </strong> - position (fbref)
 <strong>8. Min           </strong> - minutes played
 <strong>9. Gls           </strong> - goals scored by a player (fbref)
 <strong>10. Ast           </strong> - assists achieved by a player (fbref)
 <strong>11. PK            </strong> - penalty kicks made 
 <strong>12. PKatt         </strong> - penalty kicks attempted 
 <strong>13. Sh            </strong> - shots
 <strong>14. SoT           </strong> - shots on target
 <strong>15. CrdY          </strong> - yellow cards
 <strong>16. CrdR          </strong> - red cards
 <strong>17. Touches       </strong> - touches
 <strong>18. Tkl           </strong> - tackles 
 <strong>19. Int           </strong> - interceptions 
 <strong>20. Blocks        </strong> - blocks 
 <strong>21. xG            </strong> - expected goals (fbref) 
 <strong>22. npxG          </strong> - non-penalty expected goals (fbref) 
 <strong>23. xA            </strong> - expected assists (fbref) 
 <strong>24. SCA           </strong> - shot-creating actions
 <strong>25. GCA           </strong> - goal-creating actions
 <strong>26. Cmp           </strong> - passes completed
 <strong>27. Att           </strong> - passes attempted
 <strong>28. Cmp%          </strong> - pass completion %
 <strong>29. PrgP          </strong> - progressive passes
 <strong>30. Carries       </strong> - carries
 <strong>31. PrgC          </strong> - progressive carries
 <strong>32. T_att         </strong> - take-ons attempted
 <strong>33. T_succ        </strong> - successful take-ons
 <strong>34. Season        </strong> - e.g. 2022-23
 <strong>35. WDL           </strong> - win/draw/loss
 <strong>36. Team_score    </strong> - goals scored by team 
 <strong>37. Opp_score     </strong> - goals scored by opposite team 
 <strong>38. Team_CS       </strong> - did team keep a clean sheet
 <strong>39. Was_home      </strong> - was the game played home
 <strong>40. GW            </strong> - number of gameweek (fbref) 
 <strong>41. Clean_name    </strong> - name without language-specific characters
 <strong>42. FPL_name      </strong> - name (FPL)
 <strong>43. FPL_team      </strong> - team (FPL)
 <strong>44. FPL_opp_team  </strong> - opposite team (FPL) 
 <strong>45. Name_original </strong> - name with no hyphens
 <strong>46. FPL_pos       </strong> - in-game position 
 <strong>47. FPL_ast       </strong> - FPL assists
 <strong>48. Bonus         </strong> - bonus points (0-3 points)
 <strong>49. Bps                </strong> - bonus points system (FPL)
 <strong>50. Player_CS          </strong> - did player keep a clean sheet
 <strong>51. Creativity         </strong>
 <strong>52. Element            </strong>
 <strong>53. Fixture            </strong> - id of a fixture in given PL season
 <strong>54. Player_GC          </strong> - goals conceded with player on a pitch
 <strong>55. Goals_scored       </strong> - goals scored by a player (FPL)
 <strong>56. ICT_index          </strong>
 <strong>57. Influence          </strong>
 <strong>58. Kickoff_time       </strong> - YYYY-MM-DD_HH:MM:SS
 <strong>59. FPL_min            </strong> - minutes (FPL)
 <strong>60. Own_goals          </strong>
 <strong>61. Penalties_missed   </strong>
 <strong>62. Penalties_saved    </strong>
 <strong>63. Saves              </strong>
 <strong>64. Selected           </strong>
 <strong>65. Threat             </strong>
 <strong>66. FPL_points         </strong> - number of FPL points scored by a player
 <strong>67. Transfers_balance  </strong> - difference of in and out transfers
 <strong>68. Transfers_in       </strong>
 <strong>69. Transfers_out      </strong>
 <strong>70. Value              </strong> - FPL price
 <strong>71. FPL_GW             </strong> - gameweek number (FPL)
 <strong>72. xP                 </strong>
 <strong>73. FPL_xA             </strong> - expected assists (FPL) 
 <strong>74. FPL_xGI            </strong> - expected goal involvement (FPL) 
 <strong>75. FPL_xG             </strong> - expected goals (FPL) 
 <strong>76. FPL_xGC            </strong> - expected goals conceded (FPL) 
 <strong>77. Opp_rating         </strong> - opposite team elo rating at the time of the game (clubelo)
 <strong>78. Team_rating        </strong> - team elo rating at the time of the game (clubelo)

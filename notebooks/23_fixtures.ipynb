```python
def get_fixtures_for_next_gw(fixtures, gameweek):
    result = []
    for fixture in fixtures:
        if fixture['event'] == gameweek:
            result.append(fixture)
            
    return result
```

```python
def get_team_name_by_id(id, teams):
    return teams.loc[id]['name']
```

```python
all_fixtures = g.get_fixtures_data()
fixtures_next_gw = get_fixtures_for_next_gw(all_fixtures, 3)
for fixture in fixtures_next_gw:
    print(get_team_name_by_id(fixture['team_h'] - 1, teams),get_team_name_by_id(fixture['team_a'] - 1, teams))
    
```
